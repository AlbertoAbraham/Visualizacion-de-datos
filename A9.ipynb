```python
df = pd.read_csv('antropometria-datasetF.csv')
```

```python
df.head()
```

```python
modelA = sm.OLS.from_formula('cintura~peso', data = df)
resultA = modelA.fit()
resultA.summary()
```

```python
modelA = sm.OLS.from_formula('peso~cintura', data = df)
resultA = modelA.fit()
resultA.summary()
```

```python
df.corr()
```

```python
df.describe()
```

```python
resi = resultA.resid
resi.describe()
```

```python
modelB = sm.OLS.from_formula('diastol~sistol+edad+peso+cintura', data = df)
resultB = modelB.fit()
resultB.summary()
```

```python
resiB = resultB.resid
resiB.describe()
```
