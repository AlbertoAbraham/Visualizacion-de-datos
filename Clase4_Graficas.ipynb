```python
df2 = pd.read_csv('Cartwheeldata.csv')
df2.head()
```

### Existe alguna correlacion entre Wingspan vs CWDistance

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df2
          )
plt.title('Scatter plot of Wingsapn vs CWDistance', fontsize = 15)
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'CWDistance',
           data = df2,
           fit_reg = False, #si no se quiere tener la linea de regresion
           hue = 'Gender' # para ver colores por cada categoria
          )
plt.title('Scatter plot of Wingsapn vs CWDistance', fontsize = 15)
```

```python
sns.swarmplot(x = 'Gender',
              y = 'CWDistance',
              data = df2,
             )
plt.title('Swarm plot of Gender vs CWDistnace', fontsize = 15)
```

### pregunta de investigacion
#### comparaciones de wingspan de ambos generos

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1) # 1er renglon, 2da columna, 1st position
sns.boxplot(data = df2.loc[df2['Gender'] == 'F',
                               ['Wingspan']
                              ]
            )
                           
ax2 = fig.add_subplot(1,2,2) # 1er renglon, 2da columna, 2nd position
sns.boxplot(data = df2.loc[df2['Gender'] == 'M',
                               ['Wingspan']
                              ]
            )
                           
ax1.title.set_text('Genero F wingspan')
ax2.title.set_text('Genero M wingspan')
```
