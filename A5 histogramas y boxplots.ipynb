```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df.head()
```

```python
df.columns
```

```python
sns.boxplot(data = df.loc[:,['peso', 'talla', 'emb', 'temb', 'cintura', 'cadera', 'sistol',
       'diastol', 'hpresion', 'tbrazo', 'htension']])
plt.xticks(fontsize = 14, rotation =45)
plt.yticks(fontsize = 14)
plt.title('Boxplot', fontsize = 20)
```

```python
sns.distplot(df.peso)
```

```python
sns.distplot(df.talla)
```

```python
sns.distplot(df.cintura)
```

```python
sns.distplot(df.cadera)
```

```python
sns.distplot(df.sistol)
```

```python
sns.distplot(df.diastol)
```

```python
sns.lmplot(x = 'peso',
           y = 'cintura',
           data = df
          )
plt.title('diagrama disperción peso vs cintura', fontsize = 15)
```

```python
sns.boxplot(data = df.loc[:,['cintura']])
plt.xticks(fontsize = 14, rotation =45)
plt.yticks(fontsize = 14)
plt.title('Boxplot', fontsize = 20)
```

```python
plt.boxplot(df.cintura)
```

```python
sns.lineplot(data= df.loc[df['sexo'] == '1', 'cadera'])
```

```python
df2 = df.loc[df['sexo'] == '1', 'cadera']
df2.head()
```

```python
pivot1 = pd.pivot_table(df,
                        index= 'desc_ent',
                      values= 'cadera',
                      aggfunc= 'median',
                      columns= ['sexo']
                      )
df3 = pivot1
```

```python
sns.lineplot(data= df3)
```

```python
pivot2 = pd.pivot_table(df,
                        index= 'desc_ent',
                      values= 'sistol',
                      aggfunc= 'sum',
                      columns= ['sexo']
                      )
df4 = pivot2
df4
```

```python
sns.barplot(data = df4)
```

```python
sns.histplot(df.cadera, bins = 25 )
plt.show()
```

```python
sns.boxplot(data = df.loc[:,['peso']])
plt.xticks(fontsize = 14, rotation =45)
plt.yticks(fontsize = 14)
plt.title('Boxplot', fontsize = 20)
```

```python
sns.histplot(df.peso, bins = 25 )
plt.show()
```
